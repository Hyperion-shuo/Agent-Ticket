```python
from __future__ import absolute_import, division, print_function, unicode_literals

import os

import tensorflow as tf
from tensorflow import keras
```

```python
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()

train_labels = train_labels[:1000]
test_labels = test_labels[:1000]

train_images = train_images[:1000].reshape(-1, 28 * 28) / 255.0
test_images = test_images[:1000].reshape(-1, 28 * 28) / 255.0

print(train_images.shape, train_labels.shape,train_images.dtype, train_labels.dtype)
```

```python
class Model_Simple(tf.keras.Model):
    def __init__(self):
        super(Model_Simple, self).__init__()
        initializer = tf.initializers.GlorotUniform()
        regularizer = tf.keras.regularizers.l2(l=0.01)
        self.bn = tf.keras.layers.BatchNormalization(axis=1)
        self.fc1 = tf.keras.layers.Dense(200, activation='relu',
                                           kernel_initializer=initializer)
        # self.drop = tf.keras.layers.Dropout(0.2)
        self.fc2 = tf.keras.layers.Dense(10, activation="softmax"
                                         ,kernel_initializer=initializer)


    def call(self, input_tensor, training=False, show_bn=False):
        # print("is_training", training) # is_training 默认为false
        x = input_tensor
        x = self.bn(x, training)
        if show_bn:
            print("mean:",tf.math.reduce_mean(x, axis=0), "var:", tf.math.reduce_variance(x, axis=0))
        x = self.fc1(x)
        # x = self.drop(x, training)
        x = self.fc2(x)
       
        return x
# 创建一个基本的模型实例
model = Model_Simple()
model.compile(optimizer='adam',
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy'])
model(train_images[:1000], training=True, show_bn=True)
model.summary()
```

```python
model.fit(train_images, 
              train_labels,
              epochs=5, 
              validation_data=(test_images,test_labels),
              verbose=2)
```

```python
scores_call_train = model(train_images[:2],training=True)
print(scores_call_train[0],type(scores_call_train))
# model predict training 为False
scores_call_test = model(train_images[:2],training=False)
print(scores_call_test[0],type(scores_call_test))
scores_predict = model.predict(train_images[:2])
print(scores_predict[0],type(scores_predict))
```
